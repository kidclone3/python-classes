```python
def cach_1(data):
    # Merge file without sort
    ans = 0
    cur = 0
    for i in range(len(data)):
        cur = cur + data[i]
        ans = ans + cur
    return ans
```

```python
def cach_2(data):
    # Merge file with 2 pair for each time
    ans = 0
    x = data.copy()
    table = []
    while len(x) > 1:
        cur = []
        if len(x) % 2 == 1:
            cur.append(x[len(x)//2])

        for i in range(len(x)//2):
            cur.append(x[i] + x[len(x)-i-1])
            i += 1
        
        table.append(cur)
        x = cur
    ans = sum(sum(x) for x in table)
    return ans
```

```python
def cach_3(data):
    x = sorted(data)
    return cach_1(x)
```

```python
def cach_4(data):
    x = sorted(data)
    return cach_2(x)
```

```python
# create plot for 3 methods
import matplotlib.pyplot as plt
import time
import random

def plot_time(min=10, max=100, step=10):
    x = [i for i in range(min, max, step)]
    y1 = []
    y2 = []
    y3 = []
    y4 = []
    for i in x:
        data = [int(x) for x in random.sample(range(max), i)]
        start_time = time.time()
        cach_1(data)
        y1.append(time.time() - start_time)
        start_time = time.time()
        cach_2(data)
        y2.append(time.time() - start_time)
        start_time = time.time()
        cach_3(data)
        y3.append(time.time() - start_time)
        start_time = time.time()
        cach_4(data)
        y4.append(time.time() - start_time)
    plt.plot(x, y1, label='Cach 1')
    plt.plot(x, y2, label='Cach 2')
    plt.plot(x, y3, label='Cach 3')
    plt.plot(x, y4, label='Cach 4')
    plt.xlabel('Number of files')
    plt.ylabel('Time')
    plt.title('Time for 4 methods')
    plt.legend()
    plt.show()

def plot_value(min=10, max= 100, step = 10):
    x = [i for i in range(min, max, step)]
    y1 = []
    y2 = []
    y3 = []
    y4 = []
    for i in x:
        data = [int(x) for x in random.sample(range(max), i)]
        y1.append(cach_1(data))
        y2.append(cach_2(data))
        y3.append(cach_3(data))
        y4.append(cach_4(data))
    plt.plot(x, y1, label='Cach 1')
    plt.plot(x, y2, label='Cach 2')
    plt.plot(x, y3, label='Cach 3')
    plt.plot(x, y4, label='Cach 4')
    plt.xlabel('Number of files')
    plt.ylabel('Value')
    plt.title('Value for 4 methods')
    plt.legend()
    plt.show()
```

```python

plot_time(0, 10000, 1000)
plot_value(0, 10000, 1000)
```
