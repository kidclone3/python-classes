```python
import matplotlib.pyplot as plt
# extend recursion limit
import sys
sys.setrecursionlimit(100000)
```

```python
def dec2bin(x):
    if x == 0:
        return str(0)
    if x == 1:
        return str(1)
    ans = dec2bin(x//2)
    return ans + str(1) if x % 2 == 1 else ans + str(0)
```

```python
import time
```

```python
# measure run time of dec2bin and draw plot
def measure_time(func, *args):
    start = time.perf_counter()
    func(*args)
    end = time.perf_counter()
    return end - start
```

```python
# draw plot
def draw_plot_dec2bin():
    x = []
    y = []
    for i in range(1, 10000):
        x.append(i)
        y.append(measure_time(dec2bin, i))

    # create plot object
    fig, ax = plt.subplots()
    # plot has title, x label, y label
    ax.set_title('Dec2bin')
    ax.set_xlabel('n')
    ax.set_ylabel('time')
    # plot data
    ax.plot(x, y)
    # show plot
    plt.show()
    

draw_plot_dec2bin()
```

```python
def phantich(x, n):
    if x > n:
        if n != 1: return [n]
        else: return []
    
    else:
        if n % x == 0:
            while n % x == 0:
                n //= x
            return [x] + phantich(x+1, n)
        else:
            return phantich(x+1, n)
        
```

```python
# draw plot for phantich
def draw_plot_phantich():
    x = []
    y = []
    for i in range(1, 10000):
        x.append(i)
        y.append(measure_time(phantich, 2, i))

    # create plot object
    fig, ax = plt.subplots()
    # plot has title, x label, y label
    ax.set_title('Phantich')
    ax.set_xlabel('n')
    ax.set_ylabel('time')
    # plot data
    ax.plot(x, y)
    # show plot
    plt.show()

draw_plot_phantich()
```

```python
def fibo(n):
    if n == 1 or n == 2: return 1
    return fibo(n-1) + fibo(n-2)

```

```python
# draw plot for fibo
def draw_plot_fibo():
    x = []
    y = []
    for i in range(1, 30):
        x.append(i)
        y.append(measure_time(fibo, i))

    # create plot object
    fig, ax = plt.subplots()
    # plot has title, x label, y label
    ax.set_title('Fibo')
    ax.set_xlabel('n')
    ax.set_ylabel('time')
    # plot data
    ax.plot(x, y)
    # show plot
    plt.show()
draw_plot_fibo()
```
